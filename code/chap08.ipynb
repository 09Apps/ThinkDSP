```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import pandas
import scipy.signal

%precision 3
%matplotlib inline

PI2 = 2 * math.pi
```

Suppose you have a signal, like a square wave, with a lot of energy at high frequencies.

```python
signal = thinkdsp.SquareSignal(freq=440)
signal.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
```

```python
wave = signal.make_wave(duration=1, framerate=44100)
wave.make_audio()
```

One way to smooth out the high frequencies is to compute something like a local average, where you replace each point in the wave with a weighted sum of its neighbors.

One form of smoothing is a Gaussian window, which computes a weighted sum with weights from a Gaussian curve.  `scipy.signal` provides `gaussian`, which computes the weights of a Gaussian window.

```python
window = scipy.signal.gaussian(11, 2)
# window = scipy.signal.boxcar(11)
window /= sum(window)
window
```

Sure enough, the weights look like a "Bell curve", but truncated.

```python
thinkplot.plot(window)
thinkplot.config(ylim=[0, 0.25], legend=False)
```

If we pad the window so it has the same length as the wave, we can plot the window on top of the wave.

```python
segment = wave.segment(duration=0.01)
n = len(window)
padded = numpy.zeros_like(segment.ys)
padded[:n] = window
thinkplot.plot(padded)
thinkplot.plot(segment.ys)
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
```

If we multiply the window and the wave, we get the first point in the smoothed signal.

```python
prod = padded * segment.ys
sum(prod)
```

Now we can shift the window to the next location and repeat, computing a smoothed wave:

```python
convolved = numpy.zeros_like(segment.ys)
rolled = padded
for i in range(len(segment.ys)):
    convolved[i] = sum(rolled * segment.ys)
    rolled = numpy.roll(rolled, 1)
```

Here's the result, plotted on top of the original wave.

```python
wave.segment(duration=0.01).plot(color='gray')

smooth = thinkdsp.Wave(convolved, framerate=wave.framerate)
smooth.segment(duration=0.01).plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
```

Smoothing makes the square wave ramp up more slowly, and rounds off the corners a little.

This process -- applying a window to each point of a wave and computing the product -- is called convolution.  The implementation we just saw is not very efficient, but `scipy.signal` provides a better implementation:

```python
ys = numpy.convolve(wave.ys, window, mode='same')
smooth = thinkdsp.Wave(ys, framerate=wave.framerate)

wave.segment(duration=0.01).plot(color='gray')
smooth.segment(duration=0.01).plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
```

The results are similar, although `convolve` starts with the window in a different location, so the result is shifted (compared to my implementation).

```python
smooth.make_audio()
```

Smoothing makes the sound a little less "buzzy".

We can see this effect by comparing the spectrums:

```python
wave.make_spectrum().plot(color='gray')
smooth.make_spectrum().plot()
```

```python
wave.make_spectrum().amps[440]
```

```python
smooth.make_spectrum().amps[440]
```

Smoothing is a low-pass filter; it attenuates high frequencies much more than lower ones.

To quantify that effect, we can compute the ratio of the amplitudes at each frequency, before and after smoothing:

```python
amps = wave.make_spectrum().amps
amps2 = smooth.make_spectrum().amps

n = len(amps2)
amps = amps[:n]
ratio = amps2 / amps

ratio[amps<500] = 0

thinkplot.plot(ratio)
```

It looks like there is some structure to those ratios, and the convolution theorem tells us what that structure is.  Convolution with a window in the time domain has the effect of multiplying the spectrum by a filter, and the shape of the filter is the DFT of the window:

```python
n = len(window)
padded = numpy.zeros_like(wave.ys)
padded[:n] = window
fft_window = scipy.fftpack.fft(padded)
thinkplot.plot(ratio)
thinkplot.plot(abs(fft_window)[:22050])
```

It might not be obvious why the Gaussian window is a good idea.  A simple (and often used) alternative is a boxcar filter, which is a simple (unweighted) mean of consecutive samples.  To see why the boxcar window is not as good as a Gaussian, go back, replace the Gaussian window with a boxcar, and run this example again.

```python
hs = scipy.fftpack.fft(padded) * scipy.fftpack.fft(wave.ys)
thinkplot.plot(abs(hs)[:22050])
```

```python
ys = scipy.fftpack.ifft(hs)
wave2 = thinkdsp.Wave(ys.real, framerate=wave.framerate)
wave2.make_audio()
```

```python
wave.segment(duration=0.01).plot(color='gray')
wave2.segment(duration=0.01).plot()
```
