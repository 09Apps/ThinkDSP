```python
import thinkdsp
import thinkplot
import numpy
```

```python
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
wave.make_audio()
```

```python
start = 2.0
duration = 1.0
segment = wave.segment(start, duration)
segment.make_audio()
```

```python
spectrum = segment.make_spectrum()
high = 2000
spectrum.plot(high=2000)
```

```python
for f, amp in zip(spectrum.fs, spectrum.amps / 1e8):
    if amp > 0.05:
        print f, amp
```

What is a chord?

```python
freq = 220
mix = thinkdsp.TriangleSignal(freq=freq) + thinkdsp.TriangleSignal(freq=freq*1.5, amp=1.0) 
wave = mix.make_wave(duration=1, framerate=20000)
wave.normalize()
wave.apodize()
wave.make_audio()
```

```python
wave.segment(start=0.5, duration=0.021).plot()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=2500)
```
